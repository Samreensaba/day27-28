```python
import random, os, time

def diceRoll (sides):
  result = random.randint(1,sides)
  return result

def healthStat ():
  health = ((diceRoll(6) * diceRoll(12))/2) + 10
  return health

def strenghtStat():
  strength = ((diceRoll(6) * diceRoll(12))/2) + 12
  return strength

print("⚔️  BATTLE TIME  ⚔️")
print()


p1Name = input("Name your Legend: ")
p1Type = input("Character Type (Human, Elf, Wizard, Orc): ")
time.sleep(1)
print()
print(p1Name)
p1Health = healthStat()
p1Strength = strenghtStat()
print("HEALTH: ", p1Health)
print("STRENGTH: ", p1Strength)
print()
print("Who are they battling?")
print()

p2Name = input("Name your Legend: ")
p2Type = input("Character Type (Human, Elf, Wizard, Orc): ")
time.sleep(1)
print()
print(p2Name)
p2Health = healthStat()
p2Strength = strenghtStat()
print("HEALTH: ", p2Health)
print("STRENGTH: ", p2Strength)
print()

round = 1
winner = None

while True:
  time.sleep(1)
  os.system("clear")
  print("⚔️  BATTLE TIME  ⚔️")
  print()
  print("The battle begins!")
  print()
  p1Dice = diceRoll(6)
  p2Dice = diceRoll(6)
  difference = abs(p1Strength - p2Strength) + 1
  if p1Dice > p2Dice:
    p2Health -= difference
    if round == 1:
      print(p1Name, "wins the first blow")
    else:
      print(p1Name, " wins round", round)
      
    print(p2Name, "takes a hit, with ", difference, " damage.")
    
  
  elif p2Dice > p1Dice:
    p1Health -= difference
    if round == 1:
      print(p2Name, "wins the first blow")
    else:
      print(p2Name, " wins round", round)
    print(p1Name, "takes a hit, with ", difference, " damage")
  else:
    print("Their swords clash and the draw round", round)
  print()
  print(p1Name)
  print("HEALTH: ", p1Health)
  print()
  print(p2Name)
  print("HEALTH: ", p2Health)
  print()
  
  if p1Health <= 0:
    winner = p2Name
    print("Oh no ", p1Name, " has died!")
    print()
    print(p2Name, " destroyed ", p1Name, " in ", round, " rounds!")
    break
  elif p2Health <=0:
    winner = p1Name
    print("Oh no ", p2Name, " has died!")
    print()
    print(p1Name, " destroyed ", p2Name, " in ", round, " rounds!")
    break
  else:
    print("And they're both standing for the next round!")
    time.sleep(3)
    os.system("clear")
    round += 1
    

time.sleep(2)
os.system("clear")
print("⚔️  BATTLE TIME  ⚔️")
print()
print(winner, " has won in ", round, " rounds.")
  
```
